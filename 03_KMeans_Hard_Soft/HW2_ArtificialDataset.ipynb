### C. Describe what you found from the results of B. 

#### i) Findings from changing k for hard K-Means

We can see that as the number of hyperparameter K increases, the compressed image becomes somewhat similar to the original image.This is an obvious result, as increasing k means that there are more colors that we can represent. However, as we increase K, it takes more iterations for convergence.Sometimes, it doesn't even converge. For K that is small, the color seems dark. As K increases the color gets brighter. 

#### ii) Findings for changing both parameters B and K for soft-KMeans

As we increase B, the image seems to get hard and has distinct edges. This is identical to our intuition that as the stiffness parameter B goes to infinity, 
the soft K-means algorithm becomes identical to the original hard K-means. For increasing K, the color gets more vibrant and captures most of the textures as the original image.
Generally, small K tends to have a darker image compared to the original image. Soft K means are affected by nearer images
